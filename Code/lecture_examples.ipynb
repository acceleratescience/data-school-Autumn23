# Scikit-learn transformations example

```python
from sklearn.datasets import load_digits
from sklearn.preprocessing import StandardScaler, MinMaxScaler, RobustScaler
import matplotlib.pyplot as plt
```

```python
digits = load_digits()
print(digits.keys())
```

```python
X = digits.data
y = digits.target
print(X.shape, y.shape)
```

```python
standard_scaler = StandardScaler()
minmax_scaler = MinMaxScaler()
robust_scaler = RobustScaler()

X_scaled_standard = standard_scaler.fit_transform(X)
X_scaled_minmax = minmax_scaler.fit_transform(X)
X_scaled_robust = robust_scaler.fit_transform(X)
scaled = [X, X_scaled_standard, X_scaled_minmax, X_scaled_robust]
names = ['Original', 'StandardScaler', 'MinMaxScaler', 'RobustScaler']
```

```python
fig, axs = plt.subplots(1, 4, figsize=(15, 5))

for i, scaled_x in enumerate(scaled):
    axs[i].matshow(scaled_x[0].reshape(8,8), cmap='Blues')
    axs[i].set_title(f'{names[i]}')

    for j in range(8):
        for k in range(8):
            c = str(round(scaled_x[0][j*8+k], 1))
            axs[i].text(k, j, c, va='center', ha='center')
plt.savefig('../Lecture notes/img/digits_scaled.png', dpi=300)
plt.show()
```
