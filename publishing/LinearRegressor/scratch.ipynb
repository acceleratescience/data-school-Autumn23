```python
import numpy as np
from linear_regressor import *

x = np.linspace(0, 10, 1000)
y = 2.0*x + 1.0 + np.random.randn(len(x))
data_array = np.stack((x, y), axis=1)
np.savetxt('test_data.csv', data_array.T, delimiter=',')

dataset = Dataset()
dataset.from_numpy(data_array)
dataset.test_train_split(0.5)

model = LinearRegressor(dataset)

model.train()

y_train_pred = model.predict(test=False)
y_test_pred = model.predict(test=True)

import matplotlib.pyplot as plt
fig, ax = plt.subplots()
ax.plot(x, y, ls='', c='k', lw=1, ms=2, marker='o', label='data')
ax.plot(dataset.X_train, y_train_pred, 'g', lw=4, label='train')
ax.plot(dataset.X_test, y_test_pred, 'r', lw=4, label='test')

ax.set_xlabel('x')
ax.set_ylabel('y')
ax.legend()

plt.show()
```
